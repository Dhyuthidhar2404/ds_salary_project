# Salary Parsing

```python
df.shape
```

```python
df.info()
```

```python
df = df[df['Salary Estimate'] != '-1']
df.shape
```

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary
```

```python
minus_Kd = salary.apply(lambda x: x.replace('K','').replace('$',''))
minus_Kd
```

```python
minus_Kd.unique()
```

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['hourly'].unique()
```

```python
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
df['employer_provided'].unique()
```

```python
min_hr = minus_Kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
min_hr.unique()
```

```python
df['min_Salary'] = min_hr.apply(lambda x: int(x.split('-')[0]))
df
```

```python
df['min_Salary'].dtype
```

```python
df['max_Salary'] = min_hr.apply(lambda x: int(x.split('-')[1]))
df
```

```python
df['Average_Salary'] = (df.min_Salary + df.max_Salary)/2
df
```

```python
filtered_df = df[df['Rating'] < 0]
company_names = filtered_df['Company Name'].tolist()
filtered_df
```

## Company Name text only

```python
df.loc[df['Rating'] < 0, 'Company_txt'] = df['Company Name']
df.loc[df['Rating'] >= 0, 'Company_txt'] = df['Company Name'].str[:-3]
df
```

## State Field

```python
#df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
#df['job_state'] = df['Location'].str.split(',').str[1]
df.loc[:, 'job_state'] = df['Location'].str.split(',').str[1]
df
```

## Age of the Company 

```python
df.Founded.unique()
```

```python
#if there is -1 leave it like that and get the age for remaining ones.
from datetime import datetime

current_year = datetime.now().year
df['Company_Age'] = df.Founded.apply(lambda x: x if x<1 else current_year - x)
df
```

```python
df.columns
```

```python
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis =1)
df
```

## Parsing of Job Description

```python
#python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
#r studio
df['R_ys'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
#spark
df['spark_ys'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
#aws
df['aws_ys'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
#excel
df['excel_ys'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```
